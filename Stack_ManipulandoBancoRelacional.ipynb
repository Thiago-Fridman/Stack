```
#display(df_spark)

df_spark_date = df_spark.withColumn('DataAplicacao',df_spark['DataAplicacao'].cast('date'))

display(df_spark_date)
```

```
df_spark_date_SemNull = df_spark_date.filter(df_spark_date.CategoriaVacina.isNotNull())
df_spark_TotalmenteTratado = df_spark_date_SemNull.fillna(value='1ª Dose', subset=['TipoDose'])

#display(df_spark_date_SemNull2.summary())                 
display(df_spark_TotalmenteTratado)
```

```
Df_Dim_EstadoEstabelecimento = df_spark_TotalmenteTratado.select('UFdoEstabelecimento').dropDuplicates()

Df_Dim_Cor = df_spark_TotalmenteTratado.select('Cor').dropDuplicates()

Df_Dim_FabricanteVacina = df_spark_TotalmenteTratado.select('FabricanteVacina').dropDuplicates()

#display(Df_Dim_EstadoEstabelecimento)
```

```
menor_diaCollect = df_spark_TotalmenteTratado.select(min('DataAplicacao')).collect()
maior_diaCollect = df_spark_TotalmenteTratado.select(max('DataAplicacao')).collect()

for diaMENOR in menor_diaCollect:
    menor_dia=diaMENOR[0]
    
for diaMAIOR in maior_diaCollect:
    maior_dia=diaMAIOR[0]
    
#print(menor_dia)
#print(maior_dia)

```

```
#from pyspark.sql.functions import explode, sequence, to_date

beginDate = menor_dia
endDate = maior_dia

calendario = spark.sql(f"select explode(sequence(to_date('{beginDate}'), to_date('{endDate}'), interval 1 day)) as Datas")


calendario.createOrReplaceTempView('DimCalendario')
```

```

DimCalendarioAjustado = spark.sql("""


select 
Datas
,Date_format(Datas, 'dd') as Dias
,dayofweek(datas) as DiaSemana
,CASE 
		WHEN dayofweek(datas) = 1 THEN 'Domingo'
		WHEN dayofweek(datas) = 2 THEN 'Segunda-Feira'
		WHEN dayofweek(datas) = 3 THEN 'Terça-Feira'
		WHEN dayofweek(datas) = 4 THEN 'Quarta-Feira'
		WHEN dayofweek(datas) = 5 THEN 'Quinta-Feira'
		WHEN dayofweek(datas) = 6 THEN 'Sexta-Feira'
		WHEN dayofweek(datas) = 7 THEN 'Sábado'
END AS DiaSemanaTxt

,Date_format(Datas, 'MM') as Mes
,cASE
WHEN Date_format(Datas, 'MMMM') = 'January' THEN 'Janeiro'
WHEN Date_format(Datas, 'MMMM') = 'February' THEN 'Fevereiro'
WHEN Date_format(Datas, 'MMMM') = 'March' THEN 'Março'
WHEN Date_format(Datas, 'MMMM') = 'April' THEN 'Abril'
WHEN Date_format(Datas, 'MMMM') = 'May' THEN 'Maio'
WHEN Date_format(Datas, 'MMMM') = 'June' THEN 'Junho'
WHEN Date_format(Datas, 'MMMM') = 'July' THEN 'Julho'
WHEN Date_format(Datas, 'MMMM') = 'August' THEN 'Agosto'
WHEN Date_format(Datas, 'MMMM') = 'September' THEN 'Setembro'
WHEN Date_format(Datas, 'MMMM') = 'October' THEN 'Outubro'
WHEN Date_format(Datas, 'MMMM') = 'November' THEN 'Novembro'
WHEN Date_format(Datas, 'MMMM') = 'December' THEN 'Dezembro'
END as MesText

,year(Datas) as Ano
,weekofyear(Datas) AS SemanaAno

from DimCalendario

Order by Datas asc
 
""")
 

#display(DimCalendarioAjustado)
```

```
Fato_DfVacinas = df_spark_TotalmenteTratado.\
    withColumn('Mes_Ano', date_format(df_spark_TotalmenteTratado['DataAplicacao'],'MM-yyyy'))
#display(Fato_DfVacinas)

# salva as informações no datalake
spark.conf.set('spark.sql.sources.partitionOverwriteMode', 'dynamic')

Fato_DfVacinas\
    .write.mode('overwrite')\
    .format('parquet')\
    .partitionBy('Mes_Ano')\
    .save('/mnt/consumezone/THIAGO_TESTE/Stack/Fato_DfVacinas')
```
